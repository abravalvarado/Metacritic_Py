```python
import pandas as pd
import numpy as np

games = pd.read_csv(r'C:\Users\Usuario\Desktop\Kirus\Universidad\Técnico en Análisis de Datos\TAND-06\metacritic_games.csv'
                  )

metascore_avg = games['metascore'].mean()

def get_station(date):
    month = {
        'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
        'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
    }
    day, month_str, year = date.split('-')
    month = month[month_str]

    if month in (3, 4, 5):
        return 'Spring'
    elif month in (6, 7, 8):
        return 'Summer'
    elif month in (9, 10, 11):
        return 'Autumn'
    else:
        return 'Winter'

games['season'] = games['release_date'].apply(get_station)

#2.Does the score show an increase depending on the season the game is released?

#Average rating per season
#season_pivot = games.pivot_table(index='season',values='metascore',aggfunc='mean')
#season_pivot

#How many games sorted by season received a rating higher than the average?
#higherthanmean = games[games['metascore']>=promedio_metascore].pivot_table(index=['season'],values=['metascore'],aggfunc='count').sort_values(by='metascore',ascending=False)
#higherthanmean

#Graphics
#higherthanmean.plot.pie(subplots=True, figsize=(11, 6),title='Higher than the average')
#season_pivot.plot.barh()

#Conclusions
#Looking only at the average per season we can see that there is not a very big difference between the season in which the video games are released.
#However, if we analyze the amount of games that exceeded the average score we can see that there is a slight trend in Autumn.
```
